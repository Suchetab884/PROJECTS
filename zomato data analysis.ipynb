```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
orders=pd.read_csv('Orders.csv')
```

```python
customers = pd.read_csv('Customers.csv')
```

```python
restaurants = pd.read_csv('Restaurants.csv')
```

```python
orders["OrderDate"] = pd.to_datetime(orders["Order_Date"])
```

```python
orders["OrderDate"] = pd.to_datetime(orders["Order_Date"]) # Convert 'Order_Date' to datetime
orders["Month"] = orders["OrderDate"].dt.strftime("%B") # Use the new 'OrderDate' column for extracting month
```

```python
plt.figure(figsize=(6, 3))
sns.countplot(x="Month",data=orders)
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert 'Order_Date' to datetime if it's not already
orders["Order_Date"] = pd.to_datetime(orders["Order_Date"])

# Now you can use the .dt accessor to extract the hour
orders["Hour"] = orders["Order_Date"].dt.strftime("%H")
orders["Hour"] = orders["Hour"].astype("int")
orders = orders.sort_values(by="Hour")
sns.lineplot(x="Hour", y="Order_ID", data=orders)
plt.show()
```

```python
orders["Hour"]=orders["Order_Date"].dt.strftime("%H")
orders["Hour"]=orders["Hour"].astype("int")
porders=orders.sort_values(by="Hour")
sns.lineplot(x="Hour",y="Total_Amount",data=orders,estimator="sum")
plt.show
```

```python
df=pd.merge(left=orders,right=customers,on="Customer_ID",how="inner")
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x="Customer_Location",data=df,hue="Customer_Age_Group",palette="viridis",edgecolor="black")
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```
